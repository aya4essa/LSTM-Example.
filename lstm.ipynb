```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

inputs = [1, 2, 3]

h = 0
c = 0

Wf, Wi, Wo, Wc = 0.5, 0.7, 0.6, 0.9
Uf, Ui, Uo, Uc = 0.1, 0.2, 0.3, 0.4
bf, bi, bo, bc = 0, 0, 0, 0

for x in inputs:

    f = sigmoid(Wf * x + Uf * h + bf)

    i = sigmoid(Wi * x + Ui * h + bi)

    c_tilde = tanh(Wc * x + Uc * h + bc)

    c = f * c + i * c_tilde

    o = sigmoid(Wo * x + Uo * h + bo)

    h = o * tanh(c)

    print("Input:", x)
    print("Forget Gate:", f)
    print("Input Gate:", i)
    print("Cell State:", c)
    print("Hidden State:", h)
    print("-----------")

W = 1
b = 0

prediction = W * h + b

print("Predicted next value:", prediction)
```
